### Add new columns

```python
def process_date(x):
    """
    This function extracts year, month, day, hour, minute, second from a datetime feature
    :param x : datetime columns
    :return a list of extracted year, month, day, hour, minute and second
    """
    return [
        int(x.year),
        int(x.month),
        int(x.day),
        int(x.hour),
        int(x.minute),
        int(x.second),
    ]
```

```python
def add_columns(df, column_, suffix):
    """
    This function add 6 feature : year, month, day, hour, minute,
    second to the dataset and drop the datetime feature

    :param df : input datafrale.
    :param columns_ : the datetime feature.
    :param sufix : suffix we want to add to new column's names.
    :return: final dataset.
    """

    # columns of the new features we want to add
    list_columns = [
        col + suffix
        for col in ["year_", "month_", "day_", "hour_", "minute_", "second_"]
    ]

    # create a new dataframe containing extracted year, month, day, hour, minue and second from column_
    df_date = pd.DataFrame(
        process_date(x) for x in pd.to_datetime(df[column_])
    ).set_axis(list_columns, axis=1)

    # Join the dataframe created with df_continuous_features
    df = pd.concat([df_date, df], axis=1)

    # drop creation_date_answer column
    df.drop(column_, axis=1, inplace=True)

    return df
```

```python
# Extract Year, month, day, hour, minute, second from creation_date_answer column
df_continuous_features = add_columns(
    df_continuous_features, "creation_date_answer", "asr"
)

# Extract Year, month, day, hour, minute, second from creation_date_global column
df_continuous_features = add_columns(
    df_continuous_features, "creation_date_global", "glb"
)

# Extract Year, month, day, hour, minute, second from creation_date_request column
df_continuous_features = add_columns(
    df_continuous_features, "creation_date_request", "rqt"
)
```

```python
# Here, we compute Cramer's V association index between categorical features.

def cramers_V(x, y):
    """
    This function computes the cramer's V association index

    :param x : first feature
    :param y : second feature
    :retune: the Cramer's v coeeficient
    """

    # Get the crosstab based on x and y 
    crosstab = np.array(pd.crosstab(x, y, rownames=None, colnames=None))

    # Compute the chi-square statistic and 
    # p-value for the hypothesis test of independence of 
    # the observed frequencies in the contingency table
    chi2 = chi2_contingency(crosstab)[0]
    n = np.sum(crosstab)
    phi2 = chi2 / n
    r, k = crosstab.shape
    phi2corr = max(0, phi2 - ((k - 1) * (r - 1)) / (n - 1))
    rcorr, kcorr = r - ((r - 1) ** 2) / (n - 1), k - ((k - 1) ** 2) / (n - 1)
    return np.sqrt(phi2corr / min((kcorr - 1), (rcorr - 1)))
```

```python
# List of columns
list_columns = df_categorical_features.columns.tolist()

# The correlation matrix : a lower triangular matrix
rows = []

# Loop over all columns and compute cramer's v association index
for i in range(len(list_columns)):
    col = []
    for j in range(i + 1):
        cramers = cramers_V(
            df_categorical_features[list_columns[i]],
            df_categorical_features[list_columns[j]],
        )  # Cramer's V test
        col.append(round(cramers, 2))  # Keeping of the rounded value of the Cramer's V
    col += [0 for _ in range(len(list_columns) - i - 1)]
    rows.append(col)

# Correlation matrix with zeros on the diagoal
matrix_0_diag = np.array(rows)
matrix_0_diag[np.diag_indices_from(matrix_0_diag)] = 0

# sum the lower and upper traingular correlation matrix
assoc_mat = np.array(rows).transpose() + np.array(matrix_0_diag)

# Convert it to dataframe
assoc_mat = pd.DataFrame(
    assoc_mat,
    columns=df_categorical_features.columns,
    index=df_categorical_features.columns,
)
```

```python
# Plot the correlation matrix between features
plt.subplots(figsize=(15,15))  
sns.heatmap(data=assoc_mat, annot=True, cmap="BrBG")
plt.show()
```

```python
# Select upper triangle of association matrix
sup = assoc_mat.where(np.triu(np.ones(assoc_mat.shape), k=1).astype(np.bool))

# Find labels having an association greater or equal to 0.85 (Strong association).
to_drop = [column for column in sup.columns if any(sup[column] >= 0.85)]
to_drop
```

## Encode categorical features 

In this section, we'll encode categorical features :

-  For feature with hight number of unique values, we'll use Target encoding to  avoid large sparse matrices : a method that doesn’t add to the dimensionality of the dataset. In our case we have : id_group (with 15844 unique values) and location (with 99 unique values) and  fruit_situation_id (with 22 unique values)  

- Using target encoding for all categorical features may lead to model overfitting, because of its dependency to the distribution of the target.

- For the rest of the categorical features, we'll use one hot encoding :  a representation of categorical variables as binary vectors

### Traget encoding 

```python
# Categorical features with high and low frequencies
features_high_frequency = ["id_group", "fruit_situation_id", "location"]
features_low_frequency = ["AP", "gc_label", "hobby", "green_vegetables"]

# Convert type of categorical column to string
df_categorical_features[["fruit_situation_id", "location"]] = df_categorical_features[
    ["fruit_situation_id", "location"]
].astype(str)
```

```python
# Define Target encoder
te_encoder = TargetEncoder()


# Encode features_high_frequency using Target encoding
te_encoded_df = te_encoder.fit_transform(
    df_categorical_features[features_high_frequency], df_categorical_features["target"]
)

# Drop features we just encoded
df_categorical_features.drop(features_high_frequency, axis=1, inplace=True)

# Join the encoded features with the df_categorical_features
df_categorical_features = pd.concat([te_encoded_df, df_categorical_features], axis=1)

df_categorical_features.head()
```

```python
# save the one hot encoding encoder
dump(te_encoder, open('te_encoder.pkl', 'wb'))
```

### One hot encoding

```python
# creating instance of one-hot-encoder
ohe_encoder = OneHotEncoder(use_cat_names=True)

# perform one-hot encoding on features_low_frequency columns
ohe_encoder_df = pd.DataFrame(
    ohe_encoder.fit_transform(df_categorical_features[features_low_frequency])
)

# Drop features we just encoded
df_categorical_features.drop(features_low_frequency, axis=1, inplace=True)

# Join the encoded features with the df_categorical_features
df_categorical_features = pd.concat([ohe_encoder_df, df_categorical_features], axis=1)

df_categorical_features.head()
```

```python
# save the one hot encoding encoder
dump(ohe_encoder, open('ohe_encoder.pkl', 'wb'))
```

## Summurise all preprocessing steps 

```python
# Processing dataframe
def process_dataset(df):
    """
    This function applies all preporcessing steps on test dataframe :
        - Drop correlated features
        - Drop columns with missing values
        - Encode categorical features with both : Target encoder and one hot encoding

    :param  df: the test dataframe
    :return df after preprocessing
    """

    # Columns after dropping correlated features and ones with missing values
    columns_train = [
        "month_rqt",
        "day_rqt",
        "hour_rqt",
        "minute_rqt",
        "second_rqt",
        "year_glb",
        "month_glb",
        "day_glb",
        "hour_glb",
        "minute_glb",
        "second_glb",
        "situation",
        "number_of_fruit",
        "AP",
        "location",
        "gc_label",
        "id_group",
        "fruit_situation_id",
        "hobby",
        "green_vegetables",
    ]

    # Last columns used for models
    feature_model = [
        "month_rqt",
        "day_rqt",
        "hour_rqt",
        "minute_rqt",
        "second_rqt",
        "year_glb",
        "month_glb",
        "day_glb",
        "hour_glb",
        "minute_glb",
        "second_glb",
        "situation",
        "number_of_fruit",
        "AP_f",
        "AP_t",
        "gc_label_G",
        "gc_label_A",
        "gc_label_D",
        "gc_label_C",
        "gc_label_B",
        "gc_label_H",
        "gc_label_L",
        "gc_label_F",
        "gc_label_E",
        "gc_label_I",
        "gc_label_K",
        "gc_label_J",
        "hobby_football",
        "hobby_volleyball",
        "hobby_noball",
        "green_vegetables_f",
        "green_vegetables_t",
        "id_group",
        "fruit_situation_id",
        "location",
    ]

    # Categorical features with high and low frequencies
    features_high_frequency = ["id_group", "fruit_situation_id", "location"]
    features_low_frequency = ["AP", "gc_label", "hobby", "green_vegetables"]

    # Extract Year, month, day, hour, minute, second from creation_date_answer column
    df = add_columns(df, "creation_date_answer", "asr")

    # Extract Year, month, day, hour, minute, second from creation_date_global column
    df = add_columns(df, "creation_date_global", "glb")

    # Extract Year, month, day, hour, minute, second from creation_date_request column
    df = add_columns(df, "creation_date_request", "rqt")

    # Drop correlated columns and columns wih missing values,
    df = df[columns_train]

    # Convert categorical's feature type to string
    df[["fruit_situation_id", "location"]] = df[
        ["fruit_situation_id", "location"]
    ].astype(str)

    # load Target encoder
    encoder_te = pickle.load(open("te_encoder.pkl", "rb"))

    # load one hot encoder
    encoder_ohe = pickle.load(open("ohe_encoder.pkl", "rb"))

    # Apply Target encoder on features_high_frequency
    X_test_te = encoder_te.transform(df[features_high_frequency])

    # Drop features we just encoded
    df.drop(features_high_frequency, axis=1, inplace=True)

    # Join the encoded features with the df_categorical_features
    df = pd.concat([X_test_te, df], axis=1)

    # Apply the one hot encoder
    X_test_ohe = encoder_ohe.transform(df[features_low_frequency])

    # Join the encoded features with the df_categorical_features
    df = pd.concat([X_test_ohe, df], axis=1)

    return df[feature_model]
```

## Logistic Regression

```python
# Logistic Regression params
LR_params = { "multi_class": ["auto"]}

# Defining class weights
weights = {0: 1, 1: 10, 2: 100, 3: 1000}

#  Grid search for Logistic regression model
lr = LogisticRegression(class_weight=weights)
lr_tuned = GridSearchCV(
    lr,
    param_grid=LR_params,
    scoring=make_scorer(f1_score, average="macro"),
    verbose=10,
    cv=5,
    n_jobs=-1,
)

# Start training the model
lr_tuned.fit(X_train, y_train)

# Get the best estimator
lr_best = lr_tuned.best_estimator_
y_test_hat = lr_best.predict(X_test)

# Print the best parameters
print(lr_tuned.best_params_)
print(classification_report(y_test, y_test_hat))

# Plot confusion matrix
fig, ax = plt.subplots(figsize=(8, 8))
plot_confusion_matrix(
    lr_best,
    X_test,
    y_test,
    cmap=plt.cm.Blues,
    normalize="true",
    ax=ax,
)
plt.xticks(rotation="vertical")
plt.show()
```

## SVM

```python
# SVM Algorithm params
SVM_params = {"kernel": ["rbf"], "C": [1.0], "gamma": ["scale"]}

# Defining class weights
weights = {0: 1, 1: 10, 2: 100, 3: 1000}

# Start training the model
svm = SVC(class_weight=weights)
svm_tuned = GridSearchCV(
    svm,
    param_grid=SVM_params,
    scoring=make_scorer(f1_score, average="macro"),
    verbose=10,
    cv=5,
    n_jobs=-1,
)
svm_tuned.fit(X_train, y_train)

# Select the best estimator
svm_best = svm_tuned.best_estimator_
y_test_hat = svm_best.predict(X_test)

# Print the best model's parameters
print(svm_tuned.best_params_)
print(classification_report(y_test, y_test_hat))

# Plot the confusion matrix
fig, ax = plt.subplots(figsize=(8, 8))
plot_confusion_matrix(
    svm_best,
    X_test,
    y_test,
    cmap=plt.cm.Blues,
    normalize="true",
    ax=ax,
)

plt.xticks(rotation="vertical")
plt.show()
```

## Random Forest

```python
def get_weights(Y):
    """
    This function computes the penalty coefficients

    :param Y : the labels
    :return: the penalty coefficients : 1 for class 0, 10 for class 1, 100 for class 2 and 1000 for class 3
    """

    dict_weights = {0: 1, 1: 10, 2: 100, 3: 1000}
    return [dict_weights[Y.tolist()[i]] for i in range(len(Y))]
```

```python
# Random FOrest Algorithm params
RF_params = {"n_estimators": [1000], "bootstrap": [True], "n_jobs": [-1]}

# Start training the model
rdf = RandomForestClassifier()
rdf_tuned = GridSearchCV(
    rdf,
    param_grid=RF_params,
    scoring=make_scorer(f1_score, average="macro"),
    verbose=10,
    cv=5,
    n_jobs=-1,
)
rdf_tuned.fit(
    X_train,
    y_train,
    sample_weight=get_weights(y_train),
)

# Select the best estimator
rdf_best = rdf_tuned.best_estimator_
y_test_hat = rdf_best.predict(X_test)

# Plot the best model's parameters
print(rdf_tuned.best_params_)

print(classification_report(y_test, y_test_hat))
#f1_scores = f1_score(y_test, y_test_hat, average=None, labels=np.unique(y_test))
#print("----- f1 score : ", f1_scores)

# Plot the confusion matrix
fig, ax = plt.subplots(figsize=(8, 8))
plot_confusion_matrix(
    rdf_best,
    X_test,
    y_test,
    cmap=plt.cm.Blues,
    normalize="true",
    ax=ax,
)
plt.xticks(rotation="vertical")
plt.show()
```

```python
# save the random forest model
dump(rdf_best, open('rdf_best.pkl', 'wb'))
```

## Xgboost

```python
# XGBoost Algorithm params
XG_params = {
    "learning_rate": [0.02],
    "n_estimators": [1000],
    "objective": ["multi:softmax"]
    #'tree_method': ["gpu_hist"]
}

# Grid search for XGBClassifier model
xgb_tuned = GridSearchCV(
    XGBClassifier(scale_pos_weight=10),
    param_grid=XG_params,
    scoring=make_scorer(f1_score, average="macro"),
    verbose=10,
    cv=3,
    n_jobs=-1,
)

# Start training the model
xgb_tuned.fit(X_train, y_train, sample_weight=get_weights(y_train))

# Select the best estimator, and its parameters
xgb_best = xgb_tuned.best_estimator_
y_test_hat = xgb_best.predict(X_test)

# Plot the confusion matrix
print(xgb_tuned.best_params_)
print(classification_report(y_test, y_test_hat))
fig, ax = plt.subplots(figsize=(8, 8))
plot_confusion_matrix(
    xgb_best, X_test, y_test, cmap=plt.cm.Blues, normalize="true", ax=ax
)
plt.xticks(rotation="vertical")
plt.show()
```
